```python
%pip install lightgbm xgboost catboost scikit-optimize tensorflow keras
%pip install scikit-optimize

# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Machine Learning Libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import mutual_info_classif
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, matthews_corrcoef, roc_auc_score
from sklearn.linear_model import LogisticRegression
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import VotingClassifier

# Deep Learning
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Bayesian Optimization
from skopt import BayesSearchCV

# Load Dataset
df = pd.read_csv("creditcard.csv")

# Remove Duplicate Records
df = df.drop_duplicates()

# Drop Null (NaN) Values
df = df.dropna()

# Check class distribution
print("Class Distribution:\n", df["Class"].value_counts())

# Separate Features and Target
X = df.drop(columns=["Class"])
y = df["Class"]

# Standardize 'Amount' feature (Rest are PCA-transformed)
scaler = StandardScaler()
X["Amount"] = scaler.fit_transform(X[["Amount"]])

# Train-Test Split (80-20)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# ---------------- Step 1: Feature Selection using Information Gain ----------------
# Compute Information Gain (Mutual Information)
ig_values = mutual_info_classif(X_train, y_train)

# Create a DataFrame for Visualization
feature_scores = pd.DataFrame({"Feature": X.columns, "IG": ig_values})
feature_scores = feature_scores.sort_values(by="IG", ascending=False)

# Select Top 6 Features
top_features = feature_scores["Feature"].head(6).tolist()
print("Top 6 Selected Features:", top_features)

# Use Only Selected Features for Training
X_train_selected = X_train[top_features]
X_test_selected = X_test[top_features]

# ---------------- Step 2: Define Model Training Function ----------------
def train_evaluate(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Evaluation Metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])

    return {"Accuracy": accuracy, "Precision": precision, "Recall": recall, "F1 Score": f1, "MCC": mcc, "ROC_AUC": roc_auc}

# ---------------- Step 3: Train Individual Models ----------------
# Logistic Regression
lr = LogisticRegression(class_weight="balanced", solver="liblinear")
lr_results = train_evaluate(lr, X_train_selected, X_test_selected, y_train, y_test)

# LightGBM
lgb = LGBMClassifier(class_weight="balanced", random_state=42)
lgb_results = train_evaluate(lgb, X_train_selected, X_test_selected, y_train, y_test)

# XGBoost
xgb = XGBClassifier(scale_pos_weight=len(y_train[y_train == 0]) / len(y_train[y_train == 1]), use_label_encoder=False, eval_metric="logloss", random_state=42)
xgb_results = train_evaluate(xgb, X_train_selected, X_test_selected, y_train, y_test)

# CatBoost
cat = CatBoostClassifier(auto_class_weights="Balanced", verbose=0, random_state=42)
cat_results = train_evaluate(cat, X_train_selected, X_test_selected, y_train, y_test)

# Display Results
models = {"Logistic Regression": lr_results, "LightGBM": lgb_results, "XGBoost": xgb_results, "CatBoost": cat_results}
df_results = pd.DataFrame(models).T
print("\nPerformance of Individual Models:\n", df_results)

# ---------------- Step 4: Hyperparameter Optimization with Bayesian Search ----------------
# Define Hyperparameter Space for LightGBM
param_grid = {
    "num_leaves": (20, 150),
    "max_depth": (3, 15),
    "learning_rate": (0.001, 0.2, "log-uniform"),
    "n_estimators": (50, 300)
}

# Bayesian Optimization
opt_lgb = BayesSearchCV(LGBMClassifier(), param_grid, n_iter=20, cv=3, scoring="roc_auc", random_state=42, n_jobs=-1)
opt_lgb.fit(X_train_selected, y_train)

# Best Model
best_lgb = opt_lgb.best_estimator_
best_lgb_results = train_evaluate(best_lgb, X_train_selected, X_test_selected, y_train, y_test)
print("\nOptimized LightGBM Results:\n", best_lgb_results)

# ---------------- Step 5: Majority Voting Ensemble Learning ----------------
# Ensemble Model (Majority Voting)
voting_clf = VotingClassifier(estimators=[("LightGBM", lgb), ("XGBoost", xgb), ("CatBoost", cat)], voting="soft")
ensemble_results = train_evaluate(voting_clf, X_train_selected, X_test_selected, y_train, y_test)
print("\nEnsemble Model Results:\n", ensemble_results)

# ---------------- Step 6: Deep Learning (Artificial Neural Network - ANN) ----------------
# Build Deep Learning Model
dl_model = Sequential([
    Dense(86, activation="relu", input_shape=(X_train_selected.shape[1],)),
    Dense(64, activation="relu"),
    Dropout(0.3),
    Dense(32, activation="relu"),
    Dense(1, activation="sigmoid")  # Binary Classification
])

# Compile Model
dl_model.compile(optimizer=Adam(learning_rate=0.001), loss="binary_crossentropy", metrics=["accuracy"])

# Train Model
dl_model.fit(X_train_selected, y_train, epochs=10, batch_size=1563, validation_data=(X_test_selected, y_test), verbose=1)

# Predictions
y_pred_dl = (dl_model.predict(X_test_selected) > 0.5).astype("int32").flatten()

# Evaluate
dl_results = {
    "Accuracy": accuracy_score(y_test, y_pred_dl),
    "Precision": precision_score(y_test, y_pred_dl),
    "Recall": recall_score(y_test, y_pred_dl),
    "F1 Score": f1_score(y_test, y_pred_dl),
    "MCC": matthews_corrcoef(y_test, y_pred_dl),
    "ROC_AUC": roc_auc_score(y_test, dl_model.predict(X_test_selected))
}

print("\nDeep Learning Model Results:\n", dl_results)

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_curve, auc, precision_recall_curve

# ----------------- 1. Class Distribution (Fraud vs. Non-Fraud) -----------------
plt.figure(figsize=(6,4))
sns.countplot(x=y, palette=["blue", "red"])
plt.title("Class Distribution (Fraud vs. Non-Fraud)")
plt.xlabel("Class (0 = Non-Fraud, 1 = Fraud)")
plt.ylabel("Count")
plt.xticks([0,1], ["Non-Fraud", "Fraud"])
plt.show()

# ----------------- 2. Feature Importance (Information Gain) -----------------
plt.figure(figsize=(10,5))
sns.barplot(x=feature_scores["Feature"], y=feature_scores["IG"], palette="viridis")
plt.title("Feature Importance using Information Gain (Top Features)")
plt.xlabel("Features")
plt.ylabel("Information Gain (IG)")
plt.xticks(rotation=45)
plt.show()

# ----------------- 3. ROC-AUC Curve for All Models -----------------
plt.figure(figsize=(8,6))

models = [("Logistic Regression", lr), ("LightGBM", lgb), ("XGBoost", xgb), ("CatBoost", cat)]
for name, model in models:
    fpr, tpr, _ = roc_curve(y_test, model.predict_proba(X_test_selected)[:, 1])
    plt.plot(fpr, tpr, label=f"{name} (AUC = {auc(fpr, tpr):.2f})")

plt.plot([0,1], [0,1], 'k--')  # Diagonal line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC-AUC Curve")
plt.legend()
plt.show()

# ----------------- 4. Precision-Recall Curve for All Models -----------------
plt.figure(figsize=(8,6))

for name, model in models:
    precision, recall, _ = precision_recall_curve(y_test, model.predict_proba(X_test_selected)[:, 1])
    plt.plot(recall, precision, label=f"{name}")

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.show()

# ----------------- 5. Model Performance Comparison (Bar Chart) -----------------
df_results.plot(kind="bar", figsize=(12,6), colormap="coolwarm")
plt.title("Performance Comparison of Different Models")
plt.ylabel("Score")
plt.xticks(rotation=0)
plt.legend(loc="lower right")
plt.show()
```
