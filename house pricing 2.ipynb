```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
data = pd.read_csv("housing.csv")
data
```

```python
data.info()
```

```python
data.isna().sum()/len(data)*100
```

```python
data.dropna(inplace=True)
```

```python
data['ocean_proximity'].unique()
```

```python
data['ocean_proximity'].replace(['NEAR BAY', '<1H OCEAN', 'INLAND', 'NEAR OCEAN', 'ISLAND'],[0,1,2,3,4],inplace=True)
```

```python
data.info()
```

```python
X = data.drop(columns=['median_house_value'])
y = data['median_house_value']
```

```python
scaler = StandardScaler()
scaled_x = scaler.fit_transform(X)
```

```python
scaled_x
```

```python
X_train,X_test,y_train,y_test = train_test_split(scaled_x,y,test_size=0.2,random_state=0)
```

```python
regr = DecisionTreeRegressor()
regr.fit(X_train,y_train)
y_pred = regr.predict(X_test)

print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
regr = RandomForestRegressor()
regr.fit(X_train,y_train)
y_pred = regr.predict(X_test)

print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
regr = LinearRegression()
regr.fit(X_train,y_train)
y_pred = regr.predict(X_test)

print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

## Prediction

```python
import pickle

def Predict_House_Pricing(input):

    clf = pickle.load(open("rf.pkl","rb"))
    i_scaler = pickle.load(open("input_scaler.pkl","rb"))
    o_scaler = pickle.load(open("output_scaler.pkl","rb"))

    scaled_input = i_scaler.transform([input])
    prediction = clf.predict(scaled_input)

    transformed_prediction = o_scaler.inverse_transform([prediction])

    return transformed_prediction[0][0]

Predict_House_Pricing(['-5622.23','9.88','418.0','2580.0','929.0','122.0','1296.0','18.3252',3])
```
