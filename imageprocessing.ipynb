```python
import cv2 as c
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
var=c.imread('a.jpg',1)
var=c.arrowedLine(var,(0,0),(255,255),(147,96,150),5)
```

```python
rev_var=var[:, :, ::-1]
```

```python
plt.imshow(rev_var)
```

```python
blue, green, red = c.split(var)
plt.figure(figsize=[20,5])
# Display the individual color channels
plt.subplot(141);plt.imshow(blue,cmap='gray');plt.title("blue");
plt.subplot(142);plt.imshow(green,cmap='gray');plt.title("green");
plt.subplot(143);plt.imshow(red,cmap='gray');plt.title("red");

immer=c.merge((blue, green, red))
plt.subplot(144);plt.imshow(immer[:, :, ::-1]);plt.title("org");
```

```python
dat=c.cvtColor(var,c.COLOR_BGR2HSV)
h,s,v=c.split(dat)
plt.subplot(141);plt.imshow(h,cmap='gray');plt.title("hue");
plt.subplot(142);plt.imshow(s,cmap='gray');plt.title("saturation");
plt.subplot(143);plt.imshow(v,cmap='gray');plt.title("value");
mergeds=c.merge((h,s,v))
plt.imshow(mergeds)
```

```python

hs=h+30
ss=s+20
vs=v-10
merged=c.merge((hs,ss,vs))
plt.imshow(merged)
```
